```python
import matplotlib.pyplot as plt
import ee
import geemap
import xarray as xr
import xee
```

```python
ee.Authenticate()
ee.Initialize(project = 'air-quality-466614', opt_url = 'https://earthengine-highvolume.googleapis.com')
```

```python
gaul = (ee.FeatureCollection("FAO/GAUL/2015/level2")
        .filter(ee.Filter.eq('ADM0_NAME', 'Sri Lanka')))

roi = gaul.geometry()
```

```python
Map = geemap.Map()
Map.center_object(roi, 8)
Map.addLayer(roi, {'color': 'red'}, 'Sri Lanka Boundary')
Map
```

**Daily Data**

```python
daily_data = ( ee.ImageCollection("projects/sat-io/open-datasets/GHAP/GHAP_D1K_PM25").filterDate('2022','2023'))
```

Converting Images into Multidiamentional Structure

```python
daily_dataset = xr.open_dataset(
    daily_data,
    engine='ee',
    crs='EPSG:4326',
    geometry=roi,
    scale=0.001
)
```

```python
daily_average=daily_dataset.mean(dim=['lat','lon'])*0.1
```

```python
daily_df=daily_average.to_dataframe()
```

```python
daily_df.to_csv('daily_avg_january.csv')
```

**Monthly Data**

```python
monthly_data = (ee.ImageCollection("projects/sat-io/open-datasets/GHAP/GHAP_M1K_PM25").filterDate('2018','2023'))
```

```python
monthly_dataset=xr.open_dataset(monthly_data,
                                engine='ee',
                                crs='EPSG:4326',
                                geometry=roi,
                                scale=0.001
                                )
```

```python
monthly_dataset
```

```python
monthly_mean=monthly_dataset.groupby('time.month').mean('time')
```

```python
monthly_mean.b1.plot(
    x="lon",
    y="lat",
    col="month",
    robust=True,
    cmap="OrRd",
    col_wrap=4
)

plt.savefig("Monthly_PM25.png", dpi=360, bbox_inches="tight")
```

```python
monthly_avg=monthly_dataset.mean(dim=['lat','lon'])*0.1
```

```python
monthly_df=monthly_avg.to_dataframe()
```

```python
monthly_df.plot()
```

Quarterly mean for each Year

```python
quarterly_mean = monthly_dataset.groupby([ "time.year", "time.quarter" ]).mean("time")
```

```python
quarterly_mean
```

```python
quarterly_mean.b1.plot(
    x="lon",
    y="lat",
    row="year",        # each row = one year
    col="quarter",     # each column = one quarter
    robust=True,
    cmap="OrRd",
    figsize=(16, 12)
)

plt.savefig("Quarterly_PM25_by_Year.png", dpi=360, bbox_inches="tight")
```
